## **Grafica de barras o columnas agrupadas**

Este gráfico se puede utilizar cuando se trata de comparar solamente las componentes o las frecuencias en cada modalidad también se usa cuando ambas variables son cuantitativas

**Ejemplo:** Consideremos en el siguiente cuadro las variables (X) talla medida en centímetros,      (Y) peso medida en kilogramos

||$65$|$68$|$75$|
|--|--:|--:|--:|
|$160$|80|50|40|
|$167$|0|60|55|
|$175$|0|50|40|





```r
y1=c(80,0,0)
y2=c(50,60,50)
y3=c(75,55,40)
MT1=cbind(y1,y2,y3)
MT1
xi=c("160","167","175")
yi=c("65","68","75")
```

```r
barplot(MT1,beside=T,names.arg=yi,col=c(1,2,3),legend.text = c("160","167","175"))
```

```r
barplot(t(MT1),beside=T,names.arg=xi,col=c(1,2,3),legend.text=c("65","68","75"))
```

# **Grafica de barras de componentes**

Este grafico se usa para resaltar a la vez el total y las frecuencias de cada componente en cada modalidad, también se utiliza cuando una  variable es cuantitativa y la otra cualitativa

**Ejemplo:** El siguiente cuadro resume la relación entre las edades y el grado de instrucción de un grupo de personas



||$24$|$35$|$40$|$52$|
|:--|--:|--:|--:|--:|
|**Primaria**|2|0|1|2|
|**Secundaria**|1|10|6|4|
|**Superior**|6|5|7|9|


```r
y1=c(2,1,6)
y2=c(0,10,5)
y3=c(1,6,7)
y4=c(2,4,9)
MT1=cbind(y1,y2,y3,y4)
xi=c("Primaria", "Secundaria", "SUperior")
yi=c("24","35","40","52")
MT1
```

```r
barplot(MT1,beside = F,names.arg=yi,col=c(1,2,3),legend.text=xi)
```

```r
barplot(t(MT1),beside = F,names.arg=xi,col=c(1,2,3,4),legend.text=yi)
```

# **Gráfica de Paralelepipedos**

Este grafico se utiliza generalmente cuando  los datos son cuantitativos continuos

**Ejemplo:** El siguiente cuadro resume la información respecto a las variables (X) edad de padres, (Y) número de hijos

||$1$|$2$|$3$|
|:--|--:|--:|--:|
|$[25;35>$|256|114|7|
|$[35;45>$|102|295|82|
|$[45;55>$|2|41|69|



```r
install.packages("plot3D")
```

```r
library(plot3D)
```

```r
y1=c(256,102,2)
y2=c(114,295,41)
y3=c(7,82,69)
MT1=cbind(y1,y2,y3)
MT1
```

```r
xi=c("[25;35>","[35;45>","[45;55>")
yi=c("1","2","3")
MT11=structure(MT1,dim=c(3,3),dimnames=list(xi,yi))
MT11
```

```r
hist3D(x=1:3,y=1:3,z=MT1,scale=T)
```

```r
hist3D(x=1:3,y=1:3,z = MT1, scale = T, bty = "g", theta=60,phi = 20,
       col="yellow3", border = "black", shade = 0.8, ltheta = 10,
        space = 0, ticktype = "detailed", xlab="Edad", ylab="Numero de Hijos",d=2,cex.axis = 1e-9)

text3D(x = 1:3,   y = rep(-0.4,3), z = rep(3, 3),labels = rownames(MT11),add = TRUE, adj = 0)
text3D(x = rep(3.8, 3),   y = 1:3, z = rep(0, 3),labels = colnames(MT11),add = TRUE, adj = 1)
```

## **Grafico de líneas**

Los gráficos de líneas pueden utilizarse para resaltar variaciones de los datos a través del tiempo

**Ejemplo:** Consideremos el siguiente cuadro donde se especifica el año y el nivel de educación adquirido


image.png

```r
xi=c(2005,2006,2007,2008,2009)
y1=c(25,35,23,26,32)
y2=c(40,43,49,58,73)
y3=c(70,76,84,90,112)
y4=c(100,150,200,280,198)
y5=c(23,25,28,34,45)
y6=c(34,23,22,25,26)
plot(xi,y1,type="o",col="red",ylim=c(0,300))
lines(xi,y2,type="o",col="blue")
lines(xi,y3,type="o",col="green3")
lines(xi,y4,type="o",col="purple")
lines(xi,y5,type="o",col="orange")
lines(xi,y6,type="o",col="black")

```
